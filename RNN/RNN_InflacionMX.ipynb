```python
# Crear un nuevo DataFrame desde la fila 18 en adelante (índice 17)
df = df[17:]
df
```

```python
# Asignar el nuevo encabezado (new_header) como nombres de las columnas del DataFrame
df.columns = new_header

# Mostrar las primeras filas para verificar que el encabezado se asignó correctamente
df.head()
```

```python
# Reiniciar el índice del DataFrame para que empiece desde 0
# drop=True evita que la columna antigua de índices se agregue al DataFrame
# inplace=True aplica el cambio directamente en df
df.reset_index(drop=True, inplace=True)

# Mostrar las primeras filas para verificar el resultado
df.head()
```

```python
# Convertir la columna 'Título' a tipo datetime y guardarla en una nueva columna llamada 'Fecha'
df['Fecha'] = pd.to_datetime(df['Título'])

# Mostrar las primeras filas para verificar que la conversión fue correcta
df.head()
```

```python
# Filtrar por rango de fechas (del 2014 al 2024)
# Importante: usar formato 'YYYY-MM-DD' para evitar problemas de interpretación
_df = df[(df['Fecha'] >= '2014-01-01') & (df['Fecha'] <= '2024-12-31')]

_df.head()
```

```python
X = _df.drop(columns=["Índice Nacional de Precios al consumidor, variación anual", "Título", "Fecha"])
y = _df["Índice Nacional de Precios al consumidor, variación anual"].values
```

```python
X.shape #(filas, columnas)
```

```python
# Muestra las dimensiones del vector objetivo 'y' (número de muestras)
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

# Se crea un objeto MinMaxScaler que normaliza los valores entre 0 y 1
scaler = MinMaxScaler(feature_range=(0, 1))

# Se ajustan y transforman las características X para que todos sus valores estén entre 0 y 1
X_scaled = scaler.fit_transform(X)

# Se ajusta y transforma el vector objetivo y, convirtiéndolo primero en una matriz columna
y_scaled = scaler.fit_transform(y.reshape(-1, 1))
```

```python
import numpy as np

# Tamaño de la ventana temporal: cuántos pasos atrás usará el modelo para predecir
window_size = 12

# Listas donde se guardarán las secuencias (X) y los valores objetivo (y)
X_rnn = []
y_rnn = []

# Construye secuencias temporales para una RNN/LSTM/GRU
for i in range(window_size, len(X_scaled)):
    # Toma los últimos 'window_size' pasos como entrada
    X_rnn.append(X_scaled[i - window_size:i])

    # El valor a predecir es el dato correspondiente al tiempo actual i
    y_rnn.append(y_scaled[i])

# Convierte las listas en arreglos de NumPy para usarlas en un modelo de deep learning
X_rnn, y_rnn = np.array(X_rnn), np.array(y_rnn)
```

```python
# Calcula el índice donde se hará la división 80% entrenamiento, 20% prueba.
split = int(len(X) * 0.8)

# Datos de entrenamiento: todas las secuencias y etiquetas antes del 80%
X_train, y_train = X_rnn[:split], y_rnn[:split]

# Datos de prueba: el 20% restante después del índice 'split'
X_test, y_test = X_rnn[split:], y_rnn[split:]
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
window_size  # Muestra el valor de la ventana temporal usada para crear las secuencias en la RNN
```

```python
X_train.shape[2]  # Obtiene el número de características (features) por cada paso de la secuencia en el conjunto de entrenamiento
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, SimpleRNN, GRU

# Definición del modelo RNN secuencial
model = Sequential([

  # Primera capa RNN:
  # - 60 unidades
  # - return_sequences=True para devolver la secuencia completa (necesario porque hay más RNN después)
  # - input_shape indica: (longitud de la ventana temporal, número de características)
  SimpleRNN(units=60, return_sequences=True, input_shape=(window_size, X_train.shape[2])),

  # Segunda capa RNN:
  # - 30 unidades
  # - return_sequences=True para seguir pasando secuencias a la siguiente RNN
  SimpleRNN(units=30, return_sequences=True),

  # Tercera capa RNN:
  # - 15 unidades
  # - return_sequences=False porque esta es la última capa recurrente
  #   y solo necesitamos el último estado para la predicción
  SimpleRNN(units=15, return_sequences=False),

  # Capa de salida:
  # - 1 unidad porque se predice un solo valor continuo
  Dense(units=1)
])
```

```python
from tensorflow.keras.optimizers import Adam

# Configuración y compilación del modelo
learning_rate = 0.001  # Tasa de aprendizaje para el optimizador Adam

# Creamos el optimizador Adam con la tasa de aprendizaje definida
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilamos el modelo indicando:
# - optimizer: el algoritmo que ajustará los pesos (Adam)
# - loss: la función de pérdida, en este caso el error cuadrático medio (MSE)
model.compile(optimizer=adam_optimizer, loss='mean_squared_error')
```

```python
# Entrenar el modelo
model.fit(X_train, y_train, batch_size=1, epochs=10, validation_data=(X_test, y_test))
```

```python
from tensorflow.keras.optimizers import Adam

# Configuración y compilación del modelo
learning_rate = 0.001  # Tasa de aprendizaje para el optimizador Adam

# Creamos el optimizador Adam con la tasa de aprendizaje definida
adam_optimizer = Adam(learning_rate=learning_rate)

# Compilamos el modelo indicando:
# - optimizer: el algoritmo que ajustará los pesos (Adam)
# - loss: la función de pérdida, en este caso el error cuadrático medio (MSE)
model.compile(optimizer=adam_optimizer, loss='mean_squared_error')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Make predictions on the test set
predictions = model.predict(X_test)

# Calcula el MAE
mae = mean_absolute_error(y_test, predictions)
print(f"Error Absoluto Medio (MAE): {mae}")

# Calcula el MSE
mse = mean_squared_error(y_test, predictions)
print(f"Error Cuadrático Medio (MSE): {mse}")

# Calcula el RMSE
rmse = np.sqrt(mse)
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse}")

# Calcula el Coeficiente de Determinación (R²)
r2 = r2_score(y_test, predictions)
print(f"Coeficiente de Determinación (R²): {r2}")
```
